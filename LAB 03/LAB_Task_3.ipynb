```python
# TASK 1
class environment:
  def __init__(self, traffic_lvl='Light'):
    self.traffic_lvl = traffic_lvl
  def get_percept(self):
    return 'Heavy traffic' if self.traffic_lvl == 'Heavy' else 'Light traffic'
class reflexAgent:
  def __init__(self):
    pass
  def Act(self, percept):
    if percept == 'Heavy traffic':
      return 'Extend Green Time'
    elif percept == 'Light traffic':
      return 'Normal Green'

def run_agent(env, agent):
  percept = env.get_percept()
  action = agent.Act(percept)
  print(f"Percept: {percept} --> Action: {action}")

env1 = environment()
env2 = environment('Heavy')
agent = reflexAgent()
run_agent(env1, agent)
run_agent(env2, agent)
```

```python
#TASK 2

import random
class enviroment:
  def __init__(self,stud_present='yes',light_state='Off'):
    self.stud_present = stud_present
    self.light_state = light_state
  def get_percept(self):
    return {'stud_present':self.stud_present,'light_state':self.light_state}
  def open_light(self):
    if self.light_state == 'Off':
      self.light_state = 'On'
  def close_light(self):
    if self.light_state == 'On':
      self.light_state = 'Off'
class modelbasedagent:
  def __init__(self):
    self.model={'stud_present':'NO','light_state':'Off'}
  def Act(self,percept):
    self.model.update(percept)
    if self.model['stud_present'] == 'YES' and self.model['light_state'] == 'Off':
      return 'Turn Lights ON'
    elif self.model['stud_present'] == 'NO' and self.model['light_state'] == 'On':
      return 'Turn Lights OFF'
    else :
      return 'No action'
def run_agent(envir,agent,steps):
  for i in range(steps):
    envir.stud_present = random.choice(['YES', 'NO'])
    percept = envir.get_percept()
    action=agent.Act(percept)
    print(f"Step {i+1}: Percept: {percept} --> Action: {action} --> Agent Model: {agent.model}")
    if action == 'Turn Lights ON':
      envir.open_light()
    elif action == 'Turn Lights OFF':
      envir.close_light()

envir = enviroment(stud_present='YES')
agent= modelbasedagent()
run_agent(envir,agent,5)
```

```python
# TASK 3
class enviroment:
  def __init__(self):
    self.subs = ["AI", "Math", "Physics"]
    self.done = []
  def get_percept(self):
    return self.subs, self.done
  def study_sub(self, sub):
    self.done.append(sub)

class goalbasedagent:
  def __init__(self):
    self.goal = "All Subjects Completed"

  def Act(self, percept, envir):
    subs, done = percept
    for s in subs:
      if s not in done:
        print(f"Studying {s}")
        envir.study_sub(s)
    if len(subs) == len(done):
      return "Goal Achieved"
def run_agent(envir, agent):
  percept = envir.get_percept()
  action = agent.Act(percept, envir)
  if action == "Goal Achieved":
    print("Goal Achieved: All subjects completed")

envir = enviroment()
agent = goalbasedagent()
run_agent(envir, agent)
```

```python
# TASK 4
class enviroment:
  def __init__(self, rest=None):
    if rest is None:
      rest = {'A': 7, 'B': 9}
    self.rest = rest
  def get_percept(self):
    return self.rest

class utilitybasedagent:
  def __init__(self, dist=None):
    if dist is None:
      dist = {'A': 3, 'B': 5}
    self.dist = dist
  def utility(self, rating, rest_name):
    return rating - self.dist[rest_name]
  def Act(self, percept):
    best_rest = None
    best_util = -float('inf')

    for rest_name, rating in percept.items():
      util = self.utility(rating, rest_name)
      print(f"Restraunt {rest_name} Utility = {util}")
      if util > best_util:
        best_rest = rest_name
        best_util = util

    return best_rest

def run_agent(envir, agent):
  percept = envir.get_percept()
  best_rest = agent.Act(percept)
  print(f"Selected Restraunt: {best_rest}")

agent = utilitybasedagent()
envir = enviroment()
run_agent(envir, agent)
```

```python
# TASK 5
import random
class learningagent:
    def __init__(self):
        self.q_table = {'Play': 0, 'Rest': 0}
        self.reward_map = {'Play': 6, 'Rest': 1}
        self.alpha = 0.2
        self.actions = ['Play', 'Rest']
    def update_q(self, act):
        reward = self.reward_map[act]
        self.q_table[act] = self.q_table[act] + self.alpha * (reward - self.q_table[act])
    def run(self, steps=10):
        print("Learning Agent Started\n")
        for i in range(steps):
            act = random.choice(self.actions)
            reward = self.reward_map[act]
            self.update_q(act)
            print(f"Step {i+1}: Action {act} Reward {reward}")
        print("\nQ-table Updated")
        print(self.q_table)

agent = learningagent()
agent.run()
```

```python
# TASK 6
class environment:
    def __init__(self):
        self.room_status = {
            "a": "safe", "b": "safe", "c": "fire",
            "d": "safe", "e": "fire", "f": "safe",
            "g": "safe", "h": "safe", "j": "fire"
        }
        self.room_path = ["a", "b", "c", "d", "e", "f", "g", "h", "j"]
    def get_status(self, rm):
        return self.room_status[rm]
    def put_out_fire(self, rm):
        self.room_status[rm] = "safe"
    def show_environment(self):
        print("Environment Status:")
        for i in range(9):
            if self.room_status[self.room_path[i]] == "fire":
                print("fire", end=" ")
            else:
                print(" ", end=" ")
            if (i + 1) % 3 == 0:
                print()
        print("----------------------\n")

class firefightingrobot:
    def __init__(self):
        self.room_path = ["a", "b", "c", "d", "e", "f", "g", "h", "j"]
    def start(self, envir):
        print("Fire Fighting Robot Started\n")
        for rm in self.room_path:
            print(f"Robot entered room '{rm}'")
            state = envir.get_status(rm)
            if state == "fire":
                print("Fire found! Putting it out...")
                envir.put_out_fire(rm)
            else:
                print("Room is safe.")
            envir.show_environment()
        print("All fires have been handled!")
        print("Final Environment State:")
        envir.show_environment()

envir = environment()
bot = firefightingrobot()
bot.start(envir)
```
